```python
%matplotlib inline
from detail import Detail
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
annFile='../../json/PASCAL_trainval.json' # annotations
imgDir='../../images/' # jpeg images
```

```python
# initialize detail api for instance annotations
details = Detail(annFile)
```

```python
# display detail categories and supercategories
cats = details.loadCats(details.getCatIds())
nms=[cat['name'] for cat in cats]
print ('Categories: \n\n', ' '.join(nms))

nms = set([cat['supercategory'] for cat in cats])
print ('Supercategories: \n', ' '.join(nms))
```

```python
# get all images containing given categories, select one at random
catIds = details.getCatIds(catNms=['person']);
imgIds = details.getImgIds(catIds=catIds );
img = details.loadImgs(2009001427)[0]; # Load image metadata for 2009_001427.jpg
print(img);
```

```python
# Read and display image
I = io.imread(imgDir + img['file_name'])
plt.figure(); plt.axis('off')
plt.imshow(I)
plt.show()
```

```python
# load and display instance annotations
plt.imshow(I); plt.axis('off')
annIds = details.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = details.loadAnns(annIds)
details.showAnns(anns)
```
